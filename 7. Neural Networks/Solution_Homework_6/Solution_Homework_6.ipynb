### 1. Forward propogation

Note: Sigmoid function is given by 
\begin{equation}
\sigma(x) = \frac{1}{1+e^{-x}}
\end{equation}


Calculating the derivative of the logistic sigmoid function is given by: 

\begin{equation}
\sigma^{\prime}(x)=\frac{\partial}{\partial x}\left(\frac{1}{1+e^{-x}}\right)
\end{equation}

\begin{equation}
=\sigma(x)\left(1-\sigma(x)\right)
\end{equation}

### 3. Back propagation



#### a) Let’s update weight $w^2_{12}$  using gradient descent , It is given by 


$ w^2_{12} = w^2_{12} - \eta \frac{\partial L}{\partial w^2_{12}} $

To update the weight above, we need to compute partial derivate of loss w.r.t $w^2_{12}$. Using chain rule we can compute by:

$
\begin{align}
      \frac{\partial L}{\partial w^2_{12}} &= \frac{\partial L}{\partial y^2_1} * \frac{\partial y^2_1}{\partial w^2_{12}}
\end{align}
\\ \\
$

Lets individually calculate each of the above terms present on RHS

$
\begin{align}
    \frac{\partial L}{\partial y^2_1} = (y^2_1 - y_{true}) = (0.811 - 1) = -0.189
\end{align}
\\ \\
$


$
\begin{align}
    \frac{\partial y^2_1}{\partial w^2_{12}} &= y^2_1 (1 - y^2_1) * y^1_2 = 0.811 * (1 - 0.811) * 0.731 = 0.1120
\end{align}
\\ \\
$


So substituting above values into following equation we get :

$
\begin{align}
      \frac{\partial L}{\partial w^2_{12}} &= \frac{\partial L}{\partial y^2_1} * \frac{\partial y^2_1}{\partial w^2_{12}} = (-0.189) * 0.1120 = -0.0211
\end{align}
\\ \\
$


Finally we  have all the values to update the weight $w^2_{12}$, substituting values into following equation gives us value new weight :

$ w^2_{12} = w^2_{12} - \eta \frac{\partial L}{\partial w^2_{12}} = 1 - (0.1 * -0.0211) = 1.00211 $
